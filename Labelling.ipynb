```python
import cv2
import numpy as np
import pandas as pd
import os

# --- Configuration ---
TARGET_WIDTH = 800
TARGET_HEIGHT = 600
GRID_SIZE = 8
CELL_WIDTH = TARGET_WIDTH // GRID_SIZE   # 100 pixels
CELL_HEIGHT = TARGET_HEIGHT // GRID_SIZE # 75 pixels

# Global state
grid_labels = np.zeros((GRID_SIZE, GRID_SIZE), dtype=int)
image_file_name = "current_image.jpg"

def preprocess_image(image_path):
    """
    Loads, crops, scales, and pads the image to 800x600 without distortion.
    Applies Constraints 1a, 1b, and 1c.
    """
    img = cv2.imread(image_path)
    if img is None:
        raise FileNotFoundError(f"Image not found at {image_path}")

    h, w = img.shape[:2]
    target_ratio = TARGET_WIDTH / TARGET_HEIGHT # 4/3

    # --- 1. Aspect Ratio Check and Cropping (Constraint 1a) ---
    current_ratio = w / h
    
    # Check if ratio is non-4:3 (using a small tolerance)
    if abs(current_ratio - target_ratio) > 0.01:
        
        if current_ratio > target_ratio: # Wider than 4:3 (e.g., 16:9)
            # Crop width to maintain height and achieve 4:3
            new_w = int(h * target_ratio)
            crop_l = (w - new_w) // 2
            crop_r = w - new_w - crop_l
            img = img[:, crop_l:w-crop_r]
        else: # Taller than 4:3 (e.g., 3:4)
            # Crop height to maintain width and achieve 4:3
            new_h = int(w / target_ratio)
            crop_t = (h - new_h) // 2
            crop_b = h - new_h - crop_t
            img = img[crop_t:h-crop_b, :]
            
        h, w = img.shape[:2] # Update dimensions after cropping

    # --- 2. Scaling and Padding (Constraints 1b & 1c) ---
    
    if w > TARGET_WIDTH or h > TARGET_HEIGHT:
        # Case A: Image is larger -> Scale Down (Constraint 1b)
        scale = min(TARGET_WIDTH / w, TARGET_HEIGHT / h)
        new_w = int(w * scale)
        new_h = int(h * scale)
        
        # Resize using INTER_AREA for scaling down
        resized_img = cv2.resize(img, (new_w, new_h), interpolation=cv2.INTER_AREA)
        
    else:
        # Case B: Image is smaller -> No Scale Up (Constraint 1c)
        resized_img = img # Keep original small size
        new_w, new_h = w, h

    # Pad the image to 800x600
    top = (TARGET_HEIGHT - new_h) // 2
    bottom = TARGET_HEIGHT - new_h - top
    left = (TARGET_WIDTH - new_w) // 2
    right = TARGET_WIDTH - new_w - left
    
    # Create the final 800x600 canvas (black padding)
    final_img = cv2.copyMakeBorder(resized_img, top, bottom, left, right, 
                                   cv2.BORDER_CONSTANT, value=[0, 0, 0])
    
    return final_img


def draw_grid_and_labels(img):
    """Draws the 8x8 grid and highlights the labeled cells."""
    display_img = img.copy()

    # Draw the grid lines
    for i in range(1, GRID_SIZE):
        # Vertical lines
        x = i * CELL_WIDTH
        cv2.line(display_img, (x, 0), (x, TARGET_HEIGHT), (255, 255, 255), 1)
        # Horizontal lines
        y = i * CELL_HEIGHT
        cv2.line(display_img, (0, y), (TARGET_WIDTH, y), (255, 255, 255), 1)

    # Highlight labeled cells (value 1)
    for r in range(GRID_SIZE):
        for c in range(GRID_SIZE):
            if grid_labels[r, c] == 1:
                # Top-left and bottom-right corners of the cell
                x1, y1 = c * CELL_WIDTH, r * CELL_HEIGHT
                x2, y2 = (c + 1) * CELL_WIDTH, (r + 1) * CELL_HEIGHT
                # Draw a semi-transparent green overlay
                overlay = display_img.copy()
                cv2.rectangle(overlay, (x1, y1), (x2, y2), (0, 255, 0), -1)
                alpha = 0.3 # Transparency factor
                display_img = cv2.addWeighted(overlay, alpha, display_img, 1 - alpha, 0)
                
    return display_img

def mouse_callback(event, x, y, flags, param):
    """Handles mouse clicks to toggle cell labels."""
    global grid_labels
    
    if event == cv2.EVENT_LBUTTONDOWN:
        # Determine the grid cell index
        col = x // CELL_WIDTH
        row = y // CELL_HEIGHT

        if 0 <= row < GRID_SIZE and 0 <= col < GRID_SIZE:
            # Toggle the label: 0 -> 1, or 1 -> 0
            grid_labels[row, col] = 1 - grid_labels[row, col]
            
            # Re-draw the image immediately
            processed_img = draw_grid_and_labels(param['image'])
            cv2.imshow("Annotation Tool", processed_img)

def save_annotation(output_csv_path="annotations.csv"):
    """Converts the 8x8 grid into the required 1x64 CSV format."""
    global grid_labels, image_file_name
    
    # Flatten the 8x8 grid into a 1x64 vector (row-major: c01, c02, ...)
    flat_labels = grid_labels.flatten().tolist() 
    
    # Create the column names (c01 to c64)
    cols = [f'c{i+1:02d}' for i in range(GRID_SIZE * GRID_SIZE)]
    
    # Create the DataFrame row
    data = {'ImageFileName': [image_file_name]}
    for col, value in zip(cols, flat_labels):
        data[col] = [value]
        
    df = pd.DataFrame(data)

    # Check if CSV exists to append or write new header
    if not os.path.exists(output_csv_path):
        df.to_csv(output_csv_path, index=False)
    else:
        # Append without writing the header
        df.to_csv(output_csv_path, mode='a', header=False, index=False)
    
    print(f"Annotation saved for {image_file_name} to {output_csv_path}")


def run_annotation_tool(image_folder):
    """Main loop to iterate through images."""
    global grid_labels, image_file_name
    
    # --- Setup ---
    # Filter for image files
    image_paths = [os.path.join(image_folder, f) 
                   for f in os.listdir(image_folder) 
                   if f.lower().endswith(('.jpg', '.jpeg', '.png'))]
    
    if not image_paths:
        print(f"Error: No images found in the folder: {image_folder}")
        return

    cv2.namedWindow("Annotation Tool")
    
    print("--- Annotation Instructions ---")
    print("Click on a grid cell to toggle its label (1: contains wild-life, 0: empty).")
    print("Press 'N' or 'D' to save and go to the next image.")
    print("Press 'Q' to save and quit.")
    print("-------------------------------")

    for path in image_paths:
        try:
            image_file_name = os.path.basename(path)
            grid_labels = np.zeros((GRID_SIZE, GRID_SIZE), dtype=int)
            
            # Preprocess the image only once per iteration
            original_image = preprocess_image(path)
            
            # Set up the mouse callback (pass the original image data)
            cv2.setMouseCallback("Annotation Tool", mouse_callback, {'image': original_image})
            
            # Annotation loop for the current image
            while True:
                # Always draw the current state of labels onto the image
                labeled_image = draw_grid_and_labels(original_image)
                cv2.imshow("Annotation Tool", labeled_image)
                key = cv2.waitKey(1) & 0xFF
                
                # Command keys
                if key == ord('n') or key == ord('d'):
                    save_annotation()
                    break # Go to next image
                elif key == ord('q'):
                    save_annotation()
                    cv2.destroyAllWindows()
                    return # Quit the tool
        
        except FileNotFoundError as e:
            print(e)
            continue
        except Exception as e:
            print(f"An error occurred with {path}: {e}")
            continue

    cv2.destroyAllWindows()
    print("--- All images annotated. ---")

# --- Execution ---
if __name__ == '__main__':
    # Use the absolute path for your image folder here.
    # The 'r' before the string handles the backslashes correctly (raw string).
    image_directory = r'C:\Users\VARAD\OneDrive\Desktop\DS203\DS203-2025-S1-E7-project-images\images'
    run_annotation_tool(image_directory)
```
